### 1.1 The Pandas Series

### 1.2 Creating a Pandas Series


```python
# define the data and index as lists
temp = [33,19,15,89,11,-5,9]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

# create series
series_from_list = pd.Series(temp, days)
series_from_list
```

```python
# From a dictionary 
my_dict = {'Mon': 33,'Tue': 19,'Wed': 15,'Thu': 89,'Fri': 11,'Sat': -5,'Sun': 9}
series_from_dict = pd.Series(my_dict)
series_from_dict
```

```python
# From a numpy array
my_arr = np.linspace(0,10,15)
series_from_ndarray = pd.Series(my_arr)
series_from_ndarray
```

```python
# Vectorized Ops also work in Pandas Series
series_from_list * 2
```

```python
np.exp(series_from_list)
```

```python
# How satisfied are employees overall?
data['JobSatisfaction'].head()
```

```python
Job_satisfication_cat = {
    1: "Low",
    2: "Medium",
    3: "High",
    4: "Very High"
}
```

```python
data['JobSatisfaction'] = data['JobSatisfaction'].map(Job_satisfication_cat)
data['JobSatisfaction'].head()
```

```python
data['JobSatisfaction'].value_counts()
```

```python
data['JobSatisfaction'].value_counts(normalize=True)
```

```python
# List of employees with low level of Job Satisfaction 
data['JobSatisfaction'] == "Low"
```

```python
data.loc[data['JobSatisfaction'] == "Low"]
```

```python
data.loc[data['JobSatisfaction'] == "Low"].index
```

```python
# List of employees with low level of Job Satisfaction and Job Involvement
Job_Involvement_cat = {
    1: "Low",
    2: 'Medium',
    3: "High",
    4: "Very High"
}
data['JobInvolvement'] = data['JobInvolvement'].map(Job_Involvement_cat)
```

```python
data.loc[(data['JobSatisfaction'] == 'Low') & (data['JobInvolvement'] == 'Low')].index.values
```

```python
# Compare the employees with low and very high job satisfaction across the following variables :-
# Age
# Department
# Distance from Home


df_1 = data.loc[(data['JobSatisfaction'] == 'Low') | (data['JobSatisfaction'] == 'Very High')]
df_1.shape
```

```python
df_1['JobSatisfaction'].value_counts()
```

```python
grouped_df = df_1.groupby('JobSatisfaction')
```

```python
grouped_df.groups
```
