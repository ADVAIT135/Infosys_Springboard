```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(t1, f(t1), 'bo')

plt.subplot(2,1,2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

plt.ylabel("Ylabel")
```

#### 1.1 Examples

```python
x = np.linspace(start = -5, stop = 5, num = 150)
```

```python
fig,ax = plt.subplots(figsize=(7,4))
ax.plot(x,x, label='Linear')
ax.plot(x,x**2, label='Quadratic')
ax.plot(x,x**3, label='Cubic')
ax.plot(x,x**4, label='4th power')
ax.legend()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (7,5))

axes[0,0].set_title('Linear')
axes[0,0].plot(x,x)

axes[0,1].set_title('Quadratic')
axes[0,1].plot(x,x**2)

axes[1,0].set_title('Cubic')
axes[1,0].plot(x,x**3)

axes[1,1].set_title('4th Power')
axes[1,1].plot(x,x**4)

fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize = (14,5))

for i ,ax in enumerate(axes.flatten()):
    ax.set_title("x to the power {:d}".format(i+1))
    ax.plot(x, x**(i+1))

fig.tight_layout()
```

# [B] Seaborn

```python
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/housing-data/train.csv')
```

```python
data.shape
```

```python
data.info()
```

```python
num_var = ['SalePrice', 'LotArea','OverallQual', 'OverallCond',
          '1stFlrSF', '2ndFlrSF', 'BedroomAbvGr', 'YearBuilt']

cat_var = ['MSZoning', 'LotShape', 'Neighborhood', 'CentralAir',
          'SaleCondition', 'MoSold', 'YrSold']

housing = data[num_var + cat_var]
```

```python
housing.shape
```

### 2.1 Understanding the main variable

```python
housing['SalePrice'].describe()
```

```python
housing['SalePrice'].hist(edgecolor='black', bins=20)
```

```python
print("Skewness: {:0.3f}".format(housing['SalePrice'].skew()))
print("Kurtosis: {:0.3f}".format(housing['SalePrice'].kurt()))
```

### 2.2 Numerical values

```python
housing[num_var].describe()
```

```python
housing[num_var].hist(edgecolor='black', bins = 15, figsize = (14,5), layout = (2,4))
```

```python
housing['Age'] = housing['YrSold'] - housing['YearBuilt']
num_var.remove('YearBuilt')
num_var.append('Age')
```

```python
housing[num_var].hist(edgecolor='black', bins = 15, figsize = (14,5), layout = (2,4))
```

### 2.3 Categorical Variables

```python
housing['SaleCondition'].value_counts().plot(kind = 'bar', title = 'SaleCondition')
```

```python
fig, ax = plt.subplots(2, 4, figsize=(16,8))

for var, subplot in zip(cat_var, ax.flatten()):
    housing[var].value_counts().plot(kind='bar', ax=subplot, title = var)

fig.tight_layout()
```

```python
def identify_cat_above30(series):
    counts = series.value_counts()
    return list(counts[counts>=30].index)
```

```python
levels_to_keep = housing[cat_var].apply(identify_cat_above30, axis=0)
levels_to_keep
```

```python
for var in cat_var:
    housing = housing.loc[housing[var].isin(levels_to_keep[var])]
```

```python
housing.shape
```

```python
fig, ax = plt.subplots(2, 4, figsize=(16,8))

for var, subplot in zip(cat_var, ax.flatten()):
    housing[var].value_counts().plot(kind='bar', ax=subplot, title = var)

fig.tight_layout()
```

### 2.5 Relationship with Categorical variables

```python
sorted_df = housing.groupby('Neighborhood')['SalePrice'].median().sort_values().index.values
```

```python
fig, ax = plt.subplots(figsize = (14,5))
sns.boxplot(x='Neighborhood', y='SalePrice', data = housing, order = sorted_df , ax = ax)
plt.xticks(rotation = 'vertical')
```
