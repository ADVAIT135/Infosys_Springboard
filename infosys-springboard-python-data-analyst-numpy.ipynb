## 3.1 Coin Flips

```python
# Throwing a coin 1 time : 0 is trail, 1 heads
np.random.randint(low=0, high=2, size=1)
```

```python
np.random.randint(low=0, high=2, size=1)
```

```python
# Throwing a coin 10 times : 0 is trail, 1 heads
exp = np.random.randint(0,2, size=10)
print(exp)
print(exp.sum())
```

```python
exp = np.random.randint(0,2, size=10)
print(exp)
print(exp.sum())
```

```python
# Each column of this matrix will be one 10 - tosses simulation
coin_matrix = np.random.randint(0,2, size=(10000,10))
coin_matrix[:5,:]
```

```python
counts = coin_matrix.sum(axis=1)
print(counts[:25])
print(counts.mean())
print(np.median(counts))
print(counts.min(), counts.max())
print(counts.std())
```

```python
np.bincount(counts)
```

```python
unique_nums = np.arange(0,11) # The numbers that were observed in the counts vector
observed_times = np.bincount(counts) 
print("===========================================\n")
for n,count in zip(unique_nums, observed_times):
    print("{} heads observed {} times ({:0.1f}%)". format(n, count, 100*count/2000))
```
